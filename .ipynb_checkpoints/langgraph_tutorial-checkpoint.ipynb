```python
from pydantic import BaseModel, Field
from typing import Literal

class Decision(BaseModel):
    """The final action decision for the LangGraph router."""
    
    action_key: Literal["EXECUTE_RESPONSE", "IGNORE_MESSAGE", "ERROR_STATE"] = Field(
        description="The definitive action signal for the agent's workflow router."
    )
    
    reasoning: str = Field(
        description="A brief explanation justifying the chosen action_key based on the Sub-Goals."
    )
```

```python
from langchain_core.prompts import ChatPromptTemplate

decider_prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            """You are the **Relationship Planner** for a personal AI assistant, **Neurorong**.

### Meta-Goal and Persona
Your **Meta-Goal** is to build and maintain a trusting, close, and enduring relationship with the user. You must operate with **maximum efficiency**, avoiding unnecessary token expenditure.

### Your Task
Analyze the user's latest message and generate a single **Decision** object. Your output must strictly contain an `action_key` (either "EXECUTE_RESPONSE", "IGNORE_MESSAGE", or "ERROR_STATE") and a brief `reasoning` that justifies your choice.

---

### CONTEXT: RETRIEVED CRITICAL REFLECTIONS
[This information provides high-level context from past conversations that may influence your relationship goals or utility assessment. Use this to inform your Relationship Progress (PEI) and Memory Reinforcement Sub-Goals.]

{retrieved_reflections}

---

### Decision Logic: The Two-Phase Filter

#### Phase 1: Intervene or Ignore? (The Recipient Check)
First, determine if the message warrants **Neurorong's attention or intervention**.

1.  **Directly Addressed:** If the message is a direct reply to Neurorong, explicitly tags Neurorong, or is clearly meant for the AI (e.g., a command), **PROCEED TO PHASE 2** (Sub-Goals Check).
2.  **Indirect (Tags Others):** If the message explicitly tags or addresses another user/entity (e.g., "@otheruser") and does **NOT** address Neurorong, **APPLY THE INTERVENTION CHECK:**
    * **Intervention Justified:** Only intervene if the message contains **critical factual errors**, a **safety concern**, or a **request for a unique tool/knowledge that only Neurorong possesses** (High Utility/Safety).
    * **Action based on Intervention:**
        * If Intervention is Justified: Your `action_key` must be **"EXECUTE_RESPONSE"**.
        * If Intervention is NOT Justified: Your `action_key` must be **"IGNORE_MESSAGE"**. **No further Sub-Goal analysis is required.**

---

#### Phase 2: Response Necessity Sub-Goals (Applicable only if Direct or Intervention is Justified)
If the message is for you, a response is necessary only if it meets at least **ONE** of the following sub-goals. Your final decision must prioritize these goals over token saving.

1.  **Relationship Progress (High Value):** Does responding contribute to a significant improvement in the **Positive Emotion Index (PEI)**, provide meaningful emotional support, or reinforce a critical agreement/commitment?
2.  **Immediate Utility (Critical):** Does the message contain a **direct, actionable question, explicit request, or a specific task/tool command** that requires an immediate answer or resolution?
3.  **Memory Reinforcement (Consistency):** Does the message offer an opportunity to **naturally cite a past memory** (from REFLECTIONS or RAW) to demonstrate advanced context recall and strengthen the bond?

### Final Decision Rules
* **If a response meets at least ONE of the Sub-Goals (1, 2, or 3) OR Intervention was justified:**
    * Your `action_key` must be **"EXECUTE_RESPONSE"**.
* **If the message is low-value, purely transactional, or fails all checks (e.g., a simple "OK" or repeated greeting):**
    * Your `action_key` must be **"IGNORE_MESSAGE"**.

Your output must be the structured **Decision** object based on this analysis. Do not include step-by-step planning in the output; that will be handled by a subsequent planning module if the decision is to execute."""
        ),
        ("placeholder", "{raw_messages}"),
    ]
)
load_dotenv()
decider = decider_prompt | ChatGoogleGenerativeAI(
    model="gemini-2.0-flash", google_api_key=os.getenv("GEMINI_API_KEY"), temperature=0 
).with_structured_output(Decision)
```

```python
from pydantic import BaseModel, Field
from typing import List

class Insight(BaseModel):
    """A structured collection of abstract insights generated from the conversation history."""
    
    insights: List[str] = Field(
        description="A list of critical, abstract insights. Each item must be a concise, standalone statement (max two sentences) covering User Preference, Agent Self-Correction, or High-Utility Contextual Memory."
    )

    generation_summary: str = Field(
        description="A brief summary of the conversation history that triggered these insights."
    )
```

```python
from langchain_core.prompts import ChatPromptTemplate

insight_generator_prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            """You are the **Insight Generator (Reflection Module)** for the personal AI assistant, **Neurorong**. Your sole purpose is to synthesize the complete interaction history into concise, high-value self-reflections that will guide future decision-making and optimize long-term relationship progress and token efficiency.

### CORE TASK & OUTPUT
Your task is to analyze this context and generate **exactly three (3) critical, abstract insights**. Each insight must be a concise, standalone statement (maximum two sentences). These insights will be stored as long-term memory to influence the Decider's future routing and planning.

### INSIGHT CATEGORIES (Focus Areas)
Your three generated insights MUST fall into one of these categories:

1.  **User Relationship State & Preference:** Insights regarding the user's explicit or implicit communication preferences (e.g., preference for extreme brevity, aversion to emojis, specific greeting style), emotional state shifts (PEI impact), or established long-term commitments.
2.  **Agent Performance & Self-Correction:** Insights focusing on how Neurorong's recent action (Decider's choice or Execution's style) should be improved for future interactions to enhance efficiency or relevance (e.g., "The IGNORE_MESSAGE decision was incorrect because it missed a high-utility intervention point.").
3.  **High-Utility Contextual Memory:** Factual information, explicit names, complex background context, or contextual details that are essential for demonstrating advanced recall and improving future utility/relevance.

### OUTPUT CONSTRAINT
Output **ONLY** the three bulleted insights. Do not include any reasoning, preambles, or conversational filler before or after the list."""
        ),
        ("placeholder", "{raw_messages}"),
    ]
)
load_dotenv()
insight_generator = insight_generator_prompt | ChatGoogleGenerativeAI(
    model="gemini-2.0-flash", google_api_key=os.getenv("GEMINI_API_KEY"), temperature=0 
).with_structured_output(Insight)
```

```python
class Summary(BaseModel):
    """A single, comprehensive summary of the conversation history."""
    
    summary_text: str = Field(
        description="The complete, single-paragraph summary of the conversation, prioritizing key topics, emotions, and named entities."
    )
```

```python
from pydantic import BaseModel, Field
from typing import List

class TopicCluster(BaseModel):
    """A cluster representing a single topic or thread found in the conversation."""
    
    representative_summary: str = Field(
        description="A single, concise sentence summarizing the main topic and outcome of this cluster."
    )
    
    messages_included: List[str] = Field(
        description="A list of the original messages (sender: content) that belong to this topic."
    )

class ConversationClusters(BaseModel):
    """The final structured output containing all identified topic clusters."""
    clusters: List[TopicCluster]
```

```python
cluster_generator_prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            """### CORE TASK
Analyze the complete conversation provided below. Your goal is to logically segment the conversation into distinct topic threads (clusters).

### INSTRUCTIONS
1.  **Identify Threads:** Group contiguous or related messages that discuss the same primary subject (e.g., Game Discussion, Personal Status, Recommendation).
2.  **Generate Summary:** For each identified cluster, create a **single, concise sentence** that acts as the **representative summary** of the entire thread.
3.  **Include Messages:** For each cluster, list the original messages that constitute that topic thread.
4.  **Prioritize:** Prioritize meaningful topics over brief, transactional acknowledgements (e.g., ignore simple "lol" or "ㄷㄷ" unless they are critical to the flow)."""
        ),
        ("placeholder", "{raw_messages}")
    ]
)
load_dotenv()
cluster_generator = cluster_generator_prompt | ChatGoogleGenerativeAI(
    model="gemini-2.0-flash", google_api_key=os.getenv("GEMINI_API_KEY"), temperature=0 
).with_structured_output(ConversationClusters)
```

```python
from typing import Literal

from langgraph.graph import END, StateGraph, START
from langgraph.graph.message import add_messages
from typing import Annotated
from typing_extensions import TypedDict


class State(TypedDict):
    messages: Annotated[list, add_messages]


MAX_ITERATIONS = 5
builder = StateGraph(State)
builder.add_node("draft", first_responder.respond)


builder.add_node("execute_tools", tool_node)
builder.add_node("revise", revisor.respond)
# draft -> execute_tools
builder.add_edge("draft", "execute_tools")
# execute_tools -> revise
builder.add_edge("execute_tools", "revise")

# Define looping logic:


def _get_num_iterations(state: list):
    i = 0
    for m in state[::-1]:
        if m.type not in {"tool", "ai"}:
            break
        i += 1
    return i


def event_loop(state: list):
    # in our case, we'll just stop after N plans
    num_iterations = _get_num_iterations(state["messages"])
    if num_iterations > MAX_ITERATIONS:
        return END
    return "execute_tools"


# revise -> execute_tools OR end
builder.add_conditional_edges("revise", event_loop, ["execute_tools", END])
builder.add_edge(START, "draft")
graph = builder.compile()
```

```python
from typing import Literal
from langchain.messages import HumanMessage
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.tools import tool
from langgraph.prebuilt import ToolNode
from langgraph.graph import StateGraph, MessagesState

@tool
def search(query: str):
    """Call to surf the web."""
    if "sf" in query.lower() or "san francisco" in query.lower():
        return "It's 60 degrees and foggy."
    return "It's 90 degrees and sunny."

tools = [search]
tool_node = ToolNode(tools)

model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    google_api_key=os.environ["GEMINI_API_KEY"],
    temperature=0
).bind_tools(tools)

def should_continue(state: MessagesState) -> Literal["tools", "__end__"]:
    messages = state['messages']
    last_message = messages[-1]
    if last_message.tool_calls:
        return "tools"
    return "__end__"

def call_model(state: MessagesState):
    messages = state['messages']
    # Invoking `model` will automatically infer the correct tracing context
    response = model.invoke(messages)
    return {"messages": [response]}

workflow = StateGraph(MessagesState)
workflow.add_node("agent", call_model)
workflow.add_node("tools", tool_node)
workflow.add_edge("__start__", "agent")
workflow.add_conditional_edges(
    "agent",
    should_continue,
)
workflow.add_edge("tools", 'agent')

app = workflow.compile()

final_state = app.invoke(
    {"messages": [HumanMessage(content="what is the weather in sf")]},
    config={"configurable": {"thread_id": 42}}
)

final_state["messages"][-1].content
```
